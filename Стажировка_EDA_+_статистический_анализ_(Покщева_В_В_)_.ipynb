# Анализ наиболее востребованых навыков для специалистов IT сферы

```python
import json
def text_to_json(text):
  try:
    return json.loads(text.replace('"', '').replace("'", '"'))
  except:
    return []
marketjob_IT['key_skills_json'] = marketjob_IT['key_skills'].apply(text_to_json)

# возвращаем пустой список в случае каких-либо проблем
def get_values(json):
  try:
    out = []
    if len(json)>0:
      for i in json:
        out.append(i['name'])
    return out
  except:
    return []
marketjob_IT['key_skills_list'] = marketjob_IT['key_skills_json'].apply(get_values)
```

#  Гипотеза о виде распределения

На первоначальном этапе входе графической визуализации количественных данных, было выдвинуто предположение об отсутствии нормального распределения, также называемое распределением Гаусса или Гаусса — Лапласа.
Проверим данное предположение с помощью теста д'Агостино.

Гипотеза Н0: наблюдаемая выборка принадлежит нормальной генеральной совокупности, уровень значимости полагаем равным 0,05.

```python
import scipy
import scipy.stats as stats
import seaborn as sns
import sklearn.linear_model as lm
```

```python
new_marketjob_IT['salary_mean_to_rub'].skew() # коэффициент асимметрии
```

Положительный коэффициент ассимитрии говорит о левосторонем смещении "мокушки", т.е. преобладают меньшие значения. Коэффициент ассиметрии, у нормально распределенной выборки он равен 0 (возможны отклонения ±1)

```python
new_marketjob_IT['salary_mean_to_rub'].kurtosis() # коэффициент эксцесса
```

Положительный коэффициент эксцесса говорит об "остро вершинности", т.е. пик выше чем пик нормального распределения. Коэффициент эксцесса, у нормально распределенной выборки он равен 3 (возможны отклонения ±1)

```python
stats.normaltest(new_marketjob_IT['salary_mean_to_rub'].dropna())
```

Поскольку pvalue значительно ниже заданного уровня значимости (0,05), то гипотеза H0 отклоняется. Таким образом, наблюдаемая выборка не принадлежит нормальной генеральной совокупности
